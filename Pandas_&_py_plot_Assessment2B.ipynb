###### import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
sample1=pd.read_csv('/content/college_1.csv')
sample2=pd.read_csv('/content/college_2.csv')
```

```python
df=pd.concat([sample1,sample2])
```

```python
df.head()
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
# the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv
df1=df[df['CodeKata Score']>=15000]
df1.to_csv('Exceeded expectations.csv',index=False)
```

```python
#if 10000<codekata score<15000 (Reached_expectations.csv)
df2=df[(df['CodeKata Score']>=10000)&(df['CodeKata Score']<15000)]
df2.to_csv('Reached_expectations.csv',index=False)
```

```python
#if 7000<codekata score<10000 (Needs_Improvement.csv)
df3=df[(df['CodeKata Score']>=7000)&(df['CodeKata Score']<10000)]
df3.to_csv('Needs_Improvement.csv',index=False)
```

```python
#if codekate score < 7000 (Unsatisfactory.csv)
df4=df[(df['CodeKata Score']<7000)]
df4.to_csv('Unsatisfactory.csv',index=False)
```

###### Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)

```python
Y=[df["CodeKata Score"].mean(),df['Previous Geekions'].mean()]
X=["CodeKata Score","Previous Geekions"]
print("average of codekata score is {},average of geekions is {}".format(df["CodeKata Score"].mean(),df['Previous Geekions'].mean()))
plt.figure(figsize=(6,4))
plt.xlabel("Previous week scores")
plt.ylabel("Average of scores")
plt.title("Previous Geekions vs CodeKata Score",color='red',fontsize=20)
plt.bar(X,Y,color='blue',width=0.4)
plt.show()
```

###### Department wise codekata performence (pie chart)

```python
new_df=df.groupby('Department').mean()
label=['Computer Science and Engineering',
        'Electronics and Communication Engineering',
        'Electronics and Electrical Engineering']

plt.pie(new_df['CodeKata Score'],labels=label,autopct='%1.2f%%',shadow=True,startangle=120)
plt.axis('equal')
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
#Computer Science and Engineering
%matplotlib inline
doc1=df[df['Department']==('Computer Science and Engineering')]

doc1_df=doc1.sort_values(by=['CodeKata Score','Previous Geekions','Rising'],ascending=False).head(10)
X=doc1_df.Name
Y1=doc1_df.python
Y2=doc1_df.mysql
Y3=doc1_df['Previous Geekions']
Y4=doc1_df['CodeKata Score']
Y5=doc1_df.Rising
Y6=doc1_df.computational_thinking

#plot
plt.figure(figsize=(30,20))
plt.bar(X,Y1,color='r')
plt.bar(X,Y2,bottom=Y1,color='g')
plt.bar(X,Y3,bottom=Y1+Y2,color='y')
plt.bar(X,Y4,bottom=Y1+Y2+Y3,color='b')
plt.bar(X,Y5,bottom=Y1+Y2+Y3+Y4,color='black')
plt.bar(X,Y6,bottom=Y1+Y2+Y3+Y4+Y5,color='orange')
plt.xlabel('Toppers Name',fontsize=40)
plt.ylabel('score',fontsize=40)
plt.legend(['python','mysql','Previous Geekions','CodeKata Score','Rising','computational_thinking'])
plt.title('Department wise toppers(Computer Science and Engineering)',fontsize=40)
```
